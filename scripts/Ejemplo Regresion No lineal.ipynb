```python
# Importamos librerías
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Generar datos no lineales
np.random.seed(1)
X = 20 * np.random.rand(30, 1) -10     # 30 Valores aleatorios entre -10 y 10
y = (X**3) + (4*X**2) + 50 * np.random.randn(30, 1)  # Función cuadrática con ruido
                                       # 30 Valores de X elevada al cuadrado
                                       # mas un valor aleatorio de 0 a 5
```

```python
# Transformar las características para incluir términos polinomiales

grado_polynomial = 4
poly_features = PolynomialFeatures(degree=grado_polynomial)
X_poly = poly_features.fit_transform(X)
```

```python
# Crear y entrenar un modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X_poly, y)
```

```python
# Visualizar los datos y el modelo ajustado
plt.scatter(X, y, color='blue', label='Datos')
plt.xlabel('X')
plt.ylabel('y')

# Generar puntos para la curva ajustada
X_plot = np.linspace(-10, 10, 100).reshape(-1, 1)
X_plot_poly = poly_features.transform(X_plot)
y_pred = modelo.predict(X_plot_poly)

# Graficar la curva ajustada
plt.plot(X_plot, y_pred, color='red', label='Regresión polinómica (grado ' + str(grado_polynomial) + ')')

#plt.title('Regresión Polinómica (grado ' + str(grado_polynomial) + ')')
plt.legend()
plt.grid(True)
plt.show()
```
